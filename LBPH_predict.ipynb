```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
import csv
import pandas as pd
from scipy.ndimage import filters

# img = cv2.imread('C:\Users/nigge\Desktop\image recongnition\images/5498592/026.jpg')
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
# model = cv2.face.EigenFaceRecognizer_create()
# model = cv2.face.FisherFaceRecognizer_create()
model = cv2.face.LBPHFaceRecognizer_create()
print("loading xml now")
model.read("face_rec_LBPH_123.xml")
print("loading complete")
a = range(0, 99)
names = []
names = a
idList =[]
label = []
datas = []
name2= ['Id','Category']
test_images = os.listdir('images_test')
print('Found %d queries' % (len(test_images)))
for im in test_images:
    img = cv2.imread('images_test/'+im)

    signs = detector.detectMultiScale(img,
                                      scaleFactor=1.1,
                                      minNeighbors=5,
                                      minSize=(100, 100))
    if len(signs) > 0:
        for (x, y, w, h) in signs:
            cv2.rectangle(img,
                          (x, y), (x + w, y + h),
                          (0, 0, 255), 2)
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            roi = gray[x: x + w, y: y + h]
            roi = cv2.GaussianBlur(roi,(3,3),1.5)
            roi = cv2.equalizeHist(roi)
#             roi = filters.gaussian_filter(roi, 10)
            roi = cv2.resize(roi, (200, 200), interpolation=cv2.INTER_LINEAR)
            params = model.predict(roi)
        
            try:
                print("Label: %s, Confidence: %.2f" % (params[0], params[1]))

                idList.append(im)
                label.append(params[0])
                datas.append(['images/'+im,params[0]])

            except:
                continue
                print ("nothing")


    else:
        print im
        print('nothing')
        idList.append(im)
        label.append('None')
        # datas.append([im, 'None'])
        datas.append(['images/'+im, 0])

test = pd.DataFrame(columns=name2,data=datas)
print (test)
test.to_csv('submission_LBPH_123.csv',encoding='gbk',index=False)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
